## Контекст

Наши клиенты − колл-центры, мы хотим помочь им найти неэффективных операторов, а также предлагать самые выгодные тарифы, чтобы клиент долго жил с нами и не переплачивал. Нам необходимо провести анализ предоставленных данных и определить тех, клиентов, что переплачивают на текущих тарифах, а также тех, кто сможет сэкономить, если перейдет на другой тариф

Информация по тарифам:

Ежемесячная плата за тариф:

- `Тариф А — 5000 ₽`,
- `Тариф B — 2000 ₽`,
- `Тариф С — 1000 ₽`.

**Входящие звонки** — бесплатны на всех тарифах,
**Исходящие звонки внутри сети** — бесплатны до 2000 минут в месяц на всех тарифах. Цена сверх лимита:

- `Тариф A — 0.1 ₽/мин.`,
- `Тариф B — 0.15 ₽/мин.`,
- `Тариф C — 0.3 ₽/мин`.

**Исходящие звонки, кроме внутренних:**

- `Тариф A — 0.4 ₽/мин.`,
- `Тариф B — 0.5 ₽/мин.`,
- `Тариф C — 0.7₽/мин`.

За каждого оператора клиент платит `100 ₽` в месяц на любом тарифе — дополнительно к стоимости тарифа и цене звонков.   


Также, мы хотим провести анализ выручки нашего заказчика после оптимизации затрат клиентов и в случае снижения - оговорить предложения по ее оптимизации

````python
def add_plan(data: pd.DataFrame, tariffs_dict=None):
    """функция для предварительной обработки telecom_clients
       позволяет устанавливать тарифы через словарь"""

    # добаляем данные по тарифам, включая ежмесячный платеж
    # tariffs_dict - применить тарифную сетку с учетом изменений

    if tariffs_dict is None:

        data['month_fee'] = 5000. if data['tariff_plan'] == 'A' else \
            2000. if data['tariff_plan'] == 'B' else 1000.
        data['in_overplan'] = .10 if data['tariff_plan'] == 'A' else \
            .15 if data['tariff_plan'] == 'B' else .30

        data['out_not_internal'] = .40 if data['tariff_plan'] == 'A' else \
            .50 if data['tariff_plan'] == 'B' else .70

        data['operator_fee'] = 100.0
    else:

        data['month_fee'] = tariffs_dict['A'][0] if data['tariff_plan'] == 'A' else \
            tariffs_dict['B'][0] if data['tariff_plan'] == 'B' else tariffs_dict['C'][0]
        data['in_overplan'] = tariffs_dict['A'][1] if data['tariff_plan'] == 'A' else \
            tariffs_dict['B'][1] if data['tariff_plan'] == 'B' else tariffs_dict['C'][1]

        data['out_not_internal'] = tariffs_dict['A'][2] if data['tariff_plan'] == 'A' else \
            tariffs_dict['B'][2] if data['tariff_plan'] == 'B' else tariffs_dict['C'][2]

        data['operator_fee'] = 100.0

    # приводим дату к datetime

    data['date_start'] = pd.to_datetime(data['date_start'])

    return data
````

````python
def cost_func(data, min_limit=2000, col='service_paid'):
    """функция для нахождения общих затрат клиента
       min_limit - количество минут бесплатных
       col - название для конечного признака"""

    # отбираем исходящие внешние звонки и находим расходы на них


    out_nor_internal = (data['internal'] == 0)
    if not data.loc[out_nor_internal].empty:
        data.loc[out_nor_internal, col] = (
            data['total_call_duration'] * data['out_not_internal']

        )

    # отбираем исходящие внутренние звонки и находим расходы на них
    # свыше 2000 минут разговоров. Превышение умножаем на тариф
    # это будут расходы за эту часть тарифа

    out_internal_overplan = (
        (data['internal'] == 1)
        & (data['total_call_duration'] > min_limit)
    )
    if not data.loc[out_internal_overplan].empty:
        data.loc[out_internal_overplan, col] = (

            (data['total_call_duration'] - min_limit) * data['in_overplan']

        )

    # заполняем np.nan нулевым значением, если по (out & internal == 1)
    # нет значения - не было превышения
    data[col].fillna(0, inplace=True)
    
    # группируем по пользователю и месяцу и добавляем платеж за месяц
    # расходы за оператора
    data = (data
     .groupby(['user_id', 'month', 
               'month_fee', 'operator_fee', 'tariff_plan'], as_index=False)
     .agg({'total_call_duration':'sum', col:'sum', })
    )
    data[col] = data['month_fee'] + data['operator_fee'] + data[col]

    return data
````

````python
class CalculateTariff():

    """
    Класс для рассчета изменений для клиентов при переходе на иной тариф
    Работает с DataFrame

    Пример использования
      ----------
      calculator = CalculateTariff()
      new_df = calculator.get_swtich_profit(df_overpaid)

    Методы
      ---------
      get_switch_profit - на вход получает pd.DataFrame, на выходе получаем pd.DataFrame
                          с признаками - расходы клиента при условии перехода

      get_best - на вход получает pd.DataFrame, на выходе получаем pd.DataFrame
                 возвращает лучший тариф и флаг - 1 - текущий, 0 - предложенный лучше

      get_clients - на вход получает pd.DataFrame, на выходе получаем pd.DataFrame
                    возвращает клиентов, которым необходимо предложить новый тариф

    tariffs_dict = None - принимать словарь с тарифами при изменении тарифной сетки.

    """

    def __init__(self, tariffs_dict=None):
        self.tariffs_dict = tariffs_dict

    def get_switch_profit(self, tmp):
        # переключаем тарифы и смотрим затраты клиента
                
        data = tmp.copy()
        tariffs_list = ['A', 'B', 'C']
        for tariff in tariffs_list:
            
            # смотрим какого тарифа нет у пользователя
            self.adv_tariff_0,  self.adv_tariff_1 = self._apply_tariff(tariff)
            data.loc[data['tariff_plan'] == tariff,
                     f'test_tariff_{self.adv_tariff_0}'] = self.adv_tariff_0
            data.loc[data['tariff_plan'] == tariff,
                     f'test_tariff_{self.adv_tariff_1}'] = self.adv_tariff_1
            # рассчитываем тарифы для пользователя
            data = data.pipe(self._count_costs, tariff,
                            self.adv_tariff_0, self.adv_tariff_1)
        tmp = tmp.pipe(cost_func)
        data = (data
         .groupby(['user_id', 'month', 'tariff_plan'], as_index=False)
         .agg({'total_call_duration':'sum', 'switch_C':'sum', 
              'switch_A':'sum', 'switch_B':'sum'})
        .assign(tariff_plan = tmp['tariff_plan'])
        .assign(total_wo_operators = tmp['service_paid'])
        )
               
        data.fillna(0, inplace=True)
        return data

    def get_best(self, tmp):

        # из признаков затрат клиента по разным тарифам выбираем лучший
        data = tmp.copy()
        columns_ = [
                     'total_wo_operators',
                     'switch_C',
                     'switch_B',
                     'switch_A'
                   ]
        values = data[columns_].replace(0, np.nan)

        data['min_payment'] = values.min(axis=1)

        data['best'] = values.idxmin(axis=1)
        data['best'] = np.where(data['best'].apply(lambda x: x[-1]) != 's',
                                  data['best'].apply(lambda x: x[-1]), data['tariff_plan'])
        data['current_flag'] = np.where(
            data['best'] == data['tariff_plan'], 1, 0)
        data.rename(columns={
                       'tariff_plan': 'current_tariff',
                       'switch_C': 'payment_C',
                       'switch_B': 'payment_B',
                       'switch_A': 'payment_A',
                       'total_wo_operators': 'current_payment'},
                        inplace=True
                        )
        return data

    def get_clients(self, tmp, threshold=0):
        # смотрим клиентов, которым необходимо предложить другой тариф
        data = tmp.copy()
        
        data = data.loc[data['current_flag'] == 0]
        data['profit'] = data['current_payment'] - data['min_payment']
        data = data[['user_id',
                     'current_tariff',
                     'current_payment',
                     'best',
                     'min_payment',
                     'profit',
                     ]]
        data.rename(columns={
                       'best': 'recommended',
                       'min_payment': 'future_payments'},
                        inplace=True
                        )
        
        data = (data
                .groupby(['user_id', 'current_tariff', 'recommended'], as_index=False)
                .agg({'current_payment': 'mean', 'future_payments':'mean'})
                .assign(
                        profit =  lambda x: x['current_payment'] - x['future_payments'])
                .assign(
                
                        relative =  lambda  x: round((x['current_payment']/x['future_payments'] - 1), 2)) 
                )
        
        
        data.drop_duplicates(subset='user_id', keep='last', inplace=True)
        data = data.loc[data['relative'] >= threshold]
        return data

    def _apply_tariff(self, tariff):
        # вспомогательная функция. Проверяет тарифы, которых не было у клиента
        tariffs = set(['A', 'B', 'C'])
        tariff_plan = set(tariff)
        self.new_tariff = list(tariffs - tariff_plan)
        return self.new_tariff[0], self.new_tariff[1]

    def _count_costs(self, tmp, tariff, tariff_1, tariff_2):
        data = tmp.copy()
        cols = list()
        if self.tariffs_dict is None:

            self.tariffs_dict = {'A': [5000, .10, .40],
                                 'B': [2000, .15, .50],
                                 'C': [1000, .30, .70]}

        # рассчет затрат при разных тарифах аналогично
        # cost_func. Отбираем исходящие внешние звонки и находим расходы на них
        # также добавим плату за тариф и за операторов

        for tariff_new in [tariff_1, tariff_2]:

            col = f'switch_{tariff_new}'
            out_nor_internal = data.loc[(data['internal'] == 0)
                                        & (data['tariff_plan'] == tariff)]
            if not out_nor_internal.empty:
                data.loc[(data['internal'] == 0) & (data['tariff_plan'] == tariff),
                         col] = (
                data['total_call_duration'] * self.tariffs_dict[tariff_new][2]+
                self.tariffs_dict[tariff_new][0] + data['operator_fee'])
                 
            overplan = (data['internal'] == 1) & \
            (data['total_call_duration'] > 2000) & (
                data['tariff_plan'] == tariff)
            if not data.loc[overplan].empty:
                data.loc[overplan, col] = (
                (data['total_call_duration'] - 2000) *
                 self.tariffs_dict[tariff_new][1])
             
            if out_nor_internal.empty and not data.loc[overplan].empty:
                data.loc[overplan, col] = (
                (data['total_call_duration'] - 2000) *
                 self.tariffs_dict[tariff_new][1]+
                self.tariffs_dict[tariff_new][0] + data['operator_fee'])

            data[col].fillna(0, inplace=True)
            tmp[col] = data[col]
        return tmp
````

Теперь уберем из общего сета, который мы планируем анализировать, аномальных клиентов. При этом укажем и создадим константу `DEADLINE`, которая будем кодировать то количество дней, которые мы будем допускать, чтобы принять решение оставить месяц или не учитывать в анализе. `По-умолчанию - значение 2`

<a id='link1'></a>

````python
SECONDS_PER_MINUTE = 60
DEADLINE = 2

# отбрасываем аномальных клиентов, затем переводит продолжительности звонков в минуты
# округляем их в большую сторону, затем выделяем месяц из даты
# берем клиентов по DEADLINE (в днях до конца месяца)

df = (

    telecom_dataset.query('user_id not in @anomaly.user_id')
    .assign(call_duration=lambda x: np.ceil(x['call_duration'] / SECONDS_PER_MINUTE).astype('int'))
    .assign(total_call_duration=(lambda x: np.ceil(x['total_call_duration'] / SECONDS_PER_MINUTE)
                                 .astype('int')))
    .assign(month=lambda x: x['date'].dt.month)

)
max_month = df['month'].max()

# активные клиенты - те, кто работает в ноябре

active_users = df[df['month'] == max_month]['user_id'].unique()


if pd.Period(max_date.strftime('%Y-%m-%d')).days_in_month - max_date.day <= DEADLINE:
    df = df.query('user_id in @active_users and @max_month - 1 <= month <= @max_month')
else:
    df = df.query('user_id in @active_users and @max_month - 1 < @max_month')

df.drop('month', axis=1, inplace=True)
````

Найдем количество уникальных операторов на пользователя, зачтем - добавим месяц к общим данным и сгруппируем клиентов по `user_id`, месяцу, направлению звонка и признаку `internal`, посчитав сумму звонков и общего числа разговоров **с учетом времени ожидания** для исходящих и **без учета для входящих**, и найдем окончательные количества по использованным услугам, а также объединим с данными по стоимости тарифов, сразу посчитаем расходы на операторов, чтобы уменьшить количество признаков. Так же учтем, что у некоторых операторов был неопределен `operator_id`

````python
# lambda x: x.nunique(dropna=True) для исключения NaN. Находим количество уникальных операторов

unique_operators = (df.assign(month = lambda x: x['date'].dt.month)
    .groupby(['user_id', 'month'], as_index=False)['operator_id']
    .agg({'operator_id':lambda x: x.nunique(dropna=True)})
)

# работаем только с исходящими звонками
# добавляем месяц, группируем по 4 признакам, находим суммы звонков
# добавляем тарифы, назодим плату за операторов в месяц
# находим продолжительность 

df_rate = (

    df.loc[df['direction'] == 'out'].assign(month = lambda x: x['date'].dt.month)
    .groupby(['user_id', 'month', 'direction', 'internal'], as_index=False)
    .agg({'total_call_duration': 'sum'})
    .merge(unique_operators, how='left', on=['user_id', 'month'])
    .rename(columns={'operator_id': 'operators_cnt'})
    .merge(telecom_clients, how='left', on='user_id')
    .assign(operator_fee = lambda x: x['operator_fee'] * x['operators_cnt'])
    .drop(['date_start', 'operators_cnt'], axis=1)

)
````

Мы готовы к тому, чтобы посчитать затраты наших пользователей. Применим [функцию](#link2func) `cost_func`

````python
# отбираем исходящие внешние звонки и находим расходы на них
# отбираем исходящие внутренние звонки и находим расходы на них свыше 2000 минут разговоров
# находим итоговые расходы, сложив оплаты за минуты, плату за операторов и абоненсткую плату

df_payments = df_rate.pipe(cost_func)
````

Проверим три гипотезы:  

- средняя выручка нашего заказчика не изменится после переключения **переплачивающих** пользователей на более выгодные для них тарифы;

### Выручка нашего заказчика



$H_0$ **гипотеза**: статистически значимых различий в средней выручке нет после смены тарифа    
$H_1$ **гипотеза**: статистически значимые различий в средней выручке есть после смены тарифа   

___


````python
samples = (report
           .groupby('user_id')
           .agg({'current_payment':'mean', 'min_payment':'mean'})
          )
````

````python
sample_1 = samples['current_payment']
sample_2 = samples['min_payment']
# за порог значимости принимаем .05
alpha = 0.05
results = st.mannwhitneyu(sample_1, sample_2, alternative='two-sided')

print(f'p-значение: {results.pvalue}')

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя'
    ) 
    
# рассчитаем метрики для интепретации результатов  
mean_val = round(sample_2.mean()/ (sample_1.mean()) - 1, 3) * 100
abs_val = sample_2.mean() - sample_1.mean()
val = sample_2.sum() - sample_1.sum()
print(f'Относительные изменения после смены тарифа - {mean_val:.4} %')
print(f'Абсолютное среднего чека после смены тарифа - {abs_val:.2f} рубля')
print(f'Изменение выручки после смены тарифа - {val:.2f} рубля')
````

Обнаружили статистичнскую значимость в наших заключениях. По данным пользователям, заказчик может потерять в среднем `787.62 ₽` на клиента - или в сумме в разрезе исследуемых данных `120505.95 ₽`

Наблюдаем уменьшение среднего чека - а значит и снижение выручки. Преждем, чем оценить возможность перехода, необходимо согласовать с заказчиком данные потери и их возможность допустить с целью снижения вероятности оттока.

Проверим гипотезу

$H_0$ **гипотеза**: статистически значимых различий в средней выручке нет после изменения абонентской платы  
$H_1$ **гипотеза**: статистически значимые различий в средней выручке есть после изменения абонентской платы  

___


````python
samples_A = (report_A
           .groupby('user_id')
           .agg({'current_payment':'mean', 'min_payment':'mean'})
          )
````

````python
# отбираем данные для тестирования
sample_1 = samples_A['current_payment']
sample_2 = samples['current_payment']

# за порог значимости принимаем .05
alpha = 0.05
results = st.mannwhitneyu(sample_1, sample_2, alternative='two-sided')

print(f'p-значение: {results.pvalue}')

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя'
    ) 
    
# рассчитаем метрики для интепретации результатов  
mean_val = round(sample_2.mean()/ (sample_1.mean()) - 1, 3) * 100
abs_val = sample_2.mean() - sample_1.mean()
val = sample_2.sum() - sample_1.sum()
print(f'Относительные изменения после смены тарифа - {mean_val:.4} %')
print(f'Абсолютное среднего чека после смены тарифа - {abs_val:.2f} рубля')
print(f'Изменение выручки после смены тарифа - {val:.2f} рубля')
````

Обнаружили статистическую значимость. Но стоит отметить, что при данном подходе - потери заказчика ниже, средний чек уменьшается на `316.21 ` а выручка падает на `48380.70 ₽`, что меньше, чем в случае перехода на другие тарифы.
